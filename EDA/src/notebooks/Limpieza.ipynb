
**Dataset:** MetabolicSyndrome.csv

**Descripción:** Este conjunto de datos contiene información sobre individuos con síndrome metabólico, una condición médica compleja asociada a un conjunto de factores de riesgo para enfermedades cardiovasculares y diabetes tipo 2. Los datos incluyen medidas demográficas, clínicas y de laboratorio, así como la presencia o ausencia de síndrome metabólico.

**Periodo:**  
|Columna/Variable|Descripción|Tipo_de_Variable|Importancia inicial|Nota|
|-|-|-|-|-|
|seqn|Número de identificación sequencial del individuo|||||
|Age|Edad del individuo|||||
|Sex|Genero del individuo|||||
|Marital|Estado civil del individuo|||||
|Income|Nivel de ingresos del individuo|||||
|Race|Origen étnico o racial del individuo|||||
|WaistCirc|Medición de la circunferencia de la cintura|||||
|BMI|Índice de Masa Corporal|||||
|Albuminuria| Medición relacionada con la albúmina en la orina|||||
|UrAlbCr|Relación albúmina-creatinina en la orina.|||||
|UricAcid|Niveles de ácido úrico en la sangre|||||
|BloodGlucose|Niveles de glucosa en sangre, un indicador del riesgo de diabetes.|||||
|HDL|Niveles de colesterol de lipoproteínas de alta densidad("colesterol bueno")|||||
|Triglycerides|Niveles de triglicéridos en la sangre.|||||
|MetabolicSyndrome|Indica la presencia (1) o ausencia (0) de síndrome metabólico|||||
