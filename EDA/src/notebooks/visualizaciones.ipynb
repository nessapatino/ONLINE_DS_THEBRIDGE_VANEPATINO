```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
RUTA = "../data/"
```

```python
df_meta = pd.read_csv(RUTA + "SindromeMetabolico_Limpio.csv")
```

```python
def visualizar_hist(df, cat_col, num_col, group_size, color_dict, bins="auto"):
    unique_cats = df[cat_col].unique()
    num_cats = len(unique_cats)

    for i in range(0, num_cats, group_size):
        subset_cats = unique_cats[i:i+group_size]
        subset_df = df[df[cat_col].isin(subset_cats)]

        plt.figure(figsize=(10, 6))
        for cat in subset_cats:
            sns.histplot(
                subset_df[subset_df[cat_col] == cat][num_col],
                label=str(cat),
                bins=bins,
                element="poly",
                color=color_dict.get(cat, 'blue'),
                alpha = 0.3)

        plt.title("")
        plt.xlabel(num_col)
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()

# Luego, al llamar a tu función, debes proporcionar un diccionario de colores:
# Suponiendo que 'df' es tu DataFrame y 'estado_civil' es la columna que contiene los estados civiles
# 'num_col' es la columna numérica sobre la que quieres graficar.
# 'group_size' es el tamaño del grupo para iterar sobre las categorías.

# Define un diccionario de colores para asignar a cada estado civil


def visualizar_multi(df, directora, cat2, num1, color_dict, group_size = 3, bins = "auto"):
    """
    Función realiza un análisis trivariante generando histogramas para visualizar la distribución de los datos
    en relación con la variable directora y otras dos variables categóricas y numéricas.
    """
    col_directora = directora
    col_1 = cat2
    col_2 = num1
    diccionario_multivariante = {}
    for valor in df[col_directora].unique():
        diccionario_multivariante[valor] = df.loc[df[col_directora] == valor,[col_2,col_1]] 

    for valor,df_datos in diccionario_multivariante.items():
        print(f"Respuesta {valor}:")
        visualizar_hist(df_datos,col_1,col_2, group_size= group_size, bins = bins, color_dict = color_dict)
```

```python
color_dict = {
    'Single': "#A9C91E",    
    'Married':"royalblue" ,    
    'Divorced': '#ADA0B7' ,    
    'Widowed': '#F7AD00',  
    'Separated':'#E8555F'   
}
visualizar_multi(df_meta[df_meta["Marital"]!="Unk"], "MetabolicSyndrome", "Marital", "Age",color_dict= color_dict, group_size= 6)
```
