```python
# Imputamos la media a la columna "WaistCirc" y "BMI" porque vemos en el histograma y en las medidas de tendencia central que en los datos
# no hay mucha dispersión de valores, pero primero dejamos una copia del dataset original por si lo necesitaramos más adelante
df_meta_original = df_meta.copy()
waist_circ_media = df_meta["WaistCirc"].mean()
bmi_media = df_meta["BMI"].mean()
df_meta["WaistCirc"].fillna(waist_circ_media, inplace=True)
df_meta["BMI"].fillna(waist_circ_media, inplace=True)
```

```python
# Verificamos que no hayan duplicados
print(df_meta.duplicated().sum())
```

```python
# Hacemos dos columnas donde vamos a almacenar los nulos que están en la columna "Marital" e "Income", luego los reemplazamos por UKN y
# trabajaremos el analisis con los datos así para no perder filas ya que en las columnas más importantes no hay nulos

df_meta["Nulo_marital"]= df_meta["Marital"].isna()
df_meta["Nulo_income"]= df_meta["Income"].isna()
df_meta_con_nulos= df_meta.copy() #Dejamos una copia antes de reemplazar los nulos con ''
df_meta["Income"].fillna(0,inplace=True)
df_meta.fillna('Unk',inplace=True)
df_meta.info()
```

```python
df_meta.describe().T
```

```python
# Calculamos la cardinalidad y clasificamos las columnas
df_tipificacion = pd.DataFrame([df_meta.nunique(), df_meta.nunique()/len(df_meta) * 100, df_meta.dtypes]).T.rename(columns = {0: "Card",\
                                                                                                                                 1: "%_Card", 2: "Tipo"})
df_tipificacion
df_tipificacion["Clasificada_como"] = "Categorica" 
df_tipificacion.loc[df_tipificacion.Card == 2, "Clasificada_como"] = "Binaria"
df_tipificacion.loc[df_tipificacion["Card"] > 10, "Clasificada_como"] ="Numerica Discreta"
df_tipificacion.loc[df_tipificacion["%_Card"] > 30, "Clasificada_como"] = "Numerica Continua"
df_tipificacion
```
